### Import the necessary libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

### Load and Prepare the data

```python
data=pd.read_csv("/Users/basusharma/Downloads/Pandas/heart_failure_clinical_records_dataset.csv")
```

### Define Features (x) and Target (y)

```python
df=pd.DataFrame(data)
#print(df.head())
x=np.array([df["age"],df["anaemia"],df["creatinine_phosphokinase"],
           df["diabetes"],df["ejection_fraction"],df["high_blood_pressure"],
            df["platelets"],df["serum_creatinine"],df["serum_sodium"],df["sex"],
            df["smoking"],df["time"]]).T #.T: Transposes the array to shape it like (rows, columns)  needed for ML models.
y=np.array(df["DEATH_EVENT"])
```

##  Split the Data

#### 80% for training (x_train, y_train , 20% for testing (x_test, y_test)

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2,random_state=42)
```

## Train the model

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

## Accuracy Check

```python
acc=model.score(x_test,y_test)
print(acc)
```

## Take input from user 

```python
while True:
    age = int(input("Enter the age: "))
    anaemia = int(input("Anaemia (1 for yes, 0 for no): "))
    creatinine_phosphokinase = float(input("Creatinine Phosphokinase level: "))
    diabetes = int(input("Diabetes (1 for yes, 0 for no): "))
    ejection_fraction = int(input("Ejection Fraction (%): "))
    high_blood_pressure = int(input("High blood pressure (1 for yes, 0 for no): "))
    serum_creatinine = float(input("Serum Creatinine: "))
    serum_sodium = int(input("Serum Sodium: "))
    platelets = float(input("Number of Platelets: "))
    sex = int(input("Sex (1 for male, 0 for female): "))
    smoking = int(input("Smoking (1 for yes, 0 for no): "))
    time = int(input("Observation Time: "))

    user_input = np.array([[age, anaemia, creatinine_phosphokinase, diabetes, ejection_fraction,
                            high_blood_pressure, platelets, serum_creatinine, serum_sodium,
                            sex, smoking, time]])

    predict = model.predict(user_input)

    print("\n Prediction Result:")
    if predict[0] == 1:
        print(" Heart attack chances are high.")
    else:
        print(" No chance of heart attack.")

    choice = int(input("\nDo you want to check another patient? (1 for Yes / 0 for No): "))
    if choice == 0:
        print("Exiting the system. Stay healthy! ")
        break
```
